```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
import metpy.calc as mpcalc
from metpy.units import units
import numpy as np
import xarray as xr
from datetime import datetime
```

```python
Year=2011
Month=4
Day=26
Hour=12
Minute=0
dateTime= datetime(Year, Month, Hour, Minute)
timeStr= dateTime.strftime("%Y-%m-%d %H%M UTC")
```

```python
dsT = xr.open_dataset(f'/cfsr/data/{Year}/t.{Year}.0p5.anl.nc')
dsU = xr.open_dataset(f'/cfsr/data/{Year}/u.{Year}.0p5.anl.nc')
dsV = xr.open_dataset(f'/cfsr/data/{Year}/v.{Year}.0p5.anl.nc')
```

```python
latS, latN = 25., 45
lonW, lonE = -95., -75.
plevel= 500

latRange = np.arange(latS, latN + 10, 0.5)
lonRange = np.arange(lonW, lonE +10, 0.5)

T= dsT['t'].sel(time=dateTime, lev=plevel, lat=latRange, lon=lonRange)
U= dsU['u'].sel(time=dateTime, lev=plevel, lat=latRange, lon=lonRange)
V= dsV['v'].sel(time=dateTime, lev=plevel, lat=latRange, lon=lonRange)
```

```python
T = T.metpy.convert_units('degC')
U = U.metpy.convert_units('kts')
V = V.metpy.convert_units('kts')
```

```python
dx, dy = np.gradient(T, axis=(0,1))
adv_x= -U * dx
adv_y = -V * dy
```

```python
adv_total= adv_x + adv_y
adv_total
```

```python
proj_map = ccrs.LambertConformal(central_longitude=(lonW + lonE) / 2, central_latitude=(latS + latN) / 2)
proj_data = ccrs.PlateCarree()

fig, ax = plt.subplots(figsize=(12,8), subplot_kw={'projection': proj_map})
ax.set_extent([lonW, lonE, latS, latN])
ax.add_feature(cfeature.COASTLINE.with_scale('50m'))
ax.add_feature(cfeature.STATES.with_scale('50m'))

levels = np.linspace(-15, 15, 110)
cf = ax.contourf(T.lon, T.lat, adv_x, levels=levels, cmap='RdBu_r', extend='both', transform=proj_data)
cbar= plt.colorbar(cf, ax=ax, orientation='vertical', pad=0.05, aspect=20)
cbar.set_label('Temperature Advection ($^\circ$C/day)', fontsize=12)
               
plt.show
```
